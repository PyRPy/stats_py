```python
# pyplot stype
import matplotlib.pyplot as plt
import numpy as np
```

```python
# generate data
x = np.arange(0, 10, 0.2)
y = np.sin(x)
```

```python
# plot it
plt.plot(x, y)
plt.show()
```

```python
# plot it in oop stype
# generate plot
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y)
```

```python
# multiple figures/plots
x = np.arange(0, 10, 0.2)
y = np.sin(x)
z = np.cos(x)
```

```python
# generate fig and ax
fig, axs = plt.subplots(nrows=2, ncols=1)
axs[0].plot(x,y)
axs[0].set_ylabel('Sine')
axs[1].plot(x,z)
axs[1].set_ylabel('Cosine')
plt.show()
```
