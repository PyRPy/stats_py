```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns 
```

```python
x = np.linspace(1, 7, 50)
y = 3 + 2 * x + 1.5 * np.random.randn(len(x)) 
df = pd.DataFrame({'xData':x, 'yData':y})
sns.regplot('xData', 'yData', data =df)
plt.show()
```
