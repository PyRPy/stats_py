```python
# use poisson distribution to
# simulate 911 calls in the next
# hour, Sounds interesting!
```

```python
# the context looks like this:
# if a city gets about 6 (four) 911 calls
# in every 10 minutes, can you guess
# what's the chance that the police
# recieve 30(thirty) 911 calls
# in the next hour?
```

```python
# it is easy in python, once you
# set up the problem clearly:
from scipy.stats import poisson
```

```python
# estimate the mean
mu = (6 / 10) * 60
k = 30
```

```python
# calclate the probability
prob911 = poisson.pmf(k, mu)
```
